```python
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from qiskit.utils import algorithm_globals
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

algorithm_globals.random_seed = 123
DATA_PATH = '../data/processed/winedata.csv'
MODEL_PATH='../model/'
max_iter=100

data = pd.read_csv(DATA_PATH)
data.shape
```

```python
X_Train, X_Test, Y_Train, Y_Test = train_test_split(data.drop('target',axis=1),data['target'],test_size=0.25,random_state=algorithm_globals.random_seed)
```

```python
from sklearn.metrics import accuracy_score,precision_score, recall_score,f1_score
scarlar = StandardScaler();
X_Train = scarlar.fit_transform(X_Train);
X_Test = scarlar.fit_transform(X_Test);

svc = SVC(max_iter=10000,random_state=algorithm_globals.random_seed,verbose=True)
svc.fit(X_Train,Y_Train)
svc.score(X_Test,Y_Test)
Y_Pre = svc.predict(X_Test)
print('----------')
print(accuracy_score(Y_Test,Y_Pre))
print(precision_score(Y_Test,Y_Pre,average='weighted'))
print(recall_score(Y_Test,Y_Pre,average='weighted'))
print(f1_score(Y_Test,Y_Pre,average='weighted'))
```

```python
mlp = MLPClassifier(max_iter=100000000000000000,verbose=True,random_state=algorithm_globals.random_seed,n_iter_no_change=100)
mlp.fit(X_Train,Y_Train)
mlp.score(X_Test,Y_Test)
Y_Pre = mlp.predict(X_Test)
print('----------')
print(accuracy_score(Y_Test,Y_Pre))
print(precision_score(Y_Test,Y_Pre,average='weighted'))
print(recall_score(Y_Test,Y_Pre,average='weighted'))
print(f1_score(Y_Test,Y_Pre,average='weighted'))
```

```python
rfs = RandomForestClassifier(random_state=algorithm_globals.random_seed,verbose=True)
rfs.fit(X_Train,Y_Train)
rfs.score(X_Test,Y_Test)
Y_Pre = rfs.predict(X_Test)
print('----------')
print(accuracy_score(Y_Test,Y_Pre))
print(precision_score(Y_Test,Y_Pre,average='weighted'))
print(recall_score(Y_Test,Y_Pre,average='weighted'))
print(f1_score(Y_Test,Y_Pre,average='weighted'))
```
